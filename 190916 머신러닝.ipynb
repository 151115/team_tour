```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
import warnings
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("data/train.csv", index_col = "관광지명")
data.head()
```

```python
X = data.iloc[:,:12]
target = data.loc[:,"무장애관광"]
print(X.shape)
print(target.shape)
```

```python
X_train, X_test, target_train, target_test = train_test_split(X, target) 
print(X_train.shape)
print(X_test.shape)
print(target_train.shape)
print(target_test.shape)
```

```python
warnings.filterwarnings('ignore')
```

## 1. 무장애관광지 분류 모델

```python
for k in range(1, 100): # k = 13
    knn = KNeighborsClassifier(n_neighbors=k)
    result = cross_val_score(knn, X_train, target_train, cv = 5)
    print(k, result.mean())
```

```python
for c in [0.0001, 0.001, 0.01,1, 0.1, 10, 100, 1000, 10000]:#c = 10
    lr = LogisticRegression(C=c)
    result = cross_val_score(lr, X_train, target_train, cv=5)
    print(c, result.mean())
```

```python
for d in range(1, 50): #d = 5
    model = DecisionTreeClassifier(max_depth = d)
    result = cross_val_score(model, X_train, target_train, cv = 5)
    print(d, result.mean())
```

```python
from sklearn.ensemble import VotingClassifier
final_lr = LogisticRegression(C=10)
final_tree = DecisionTreeClassifier(max_depth=7)
voting = VotingClassifier(estimators = [('lr', final_lr),('tree', final_tree)], voting='soft')
voting.fit(X_train, target_train)
print(voting.score(X_train,target_train))
print(voting.score(X_test, target_test))
```
