```python
fin_data = pd.read_csv('data/전체_전처리_02.csv')
fin_data.head()
```

```python
fin_list = fin_data["관광지명"].values.tolist()
```

```python
physical = pd.read_csv("data/지체장애.csv")
physical_list = physical["0"].values.tolist()
physical_result = []
for i in fin_list:
    if(i in physical_list):
        physical_result.append(1)
    else:
        physical_result.append(0)
```

```python
baby = pd.read_csv("data/영유아가족.csv")
baby_list = baby["0"].values.tolist()
baby_result = []
for i in fin_list:
    if(i in baby_list):
        baby_result.append(1)
    else:
        baby_result.append(0)
```

```python
senior = pd.read_csv("data/어르신.csv")
senior_list = senior["0"].values.tolist()
senior_result = []
for i in fin_list:
    if(i in senior_list):
        senior_result.append(1)
    else:
        senior_result.append(0)
```

```python
eye = pd.read_csv("data/시각장애.csv")
eye_list = eye["0"].values.tolist()
eye_result = []
for i in fin_list:
    if(i in eye_list):
        eye_result.append(1)
    else:
        eye_result.append(0)
```

```python
ear = pd.read_csv("data/청각장애.csv")
ear_list = ear["0"].values.tolist()
ear_result = []
for i in fin_list:
    if(i in ear_list):
        ear_result.append(1)
    else:
        ear_result.append(0)
```

```python
free_result = []
for i in range(len(physical_result)):
    if(physical_result[i] == 1 and senior_result[i]==1 and baby_result[i] == 1):
        free_result.append(1)
    else:
        free_result.append(0)
```

```python
barrierfree_df = pd.DataFrame({"지체장애":physical_result, "영유아가족":baby_result, "어르신": senior_result, "시각장애":eye_result, "청각장애":ear_result ,"무장애관광":free_result})
barrierfree_df.head()
```

```python
barrierfree_df["무장애관광"].value_counts()
```

```python
fin_data = pd.concat([fin_data, barrierfree_df],axis = 1)
```

```python
fin_data.set_index("관광지명", inplace=True)
```

```python
fin_data.to_csv("data/train.csv", index = "관광지명")
```
