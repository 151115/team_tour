```python
import pandas as pd
data2 = pd.read_csv("data/train.csv")
```

```python
data2
```

```python
list1 = data2["지체장애"].values.tolist()
list2 = data2["영유아가족"].values.tolist()
```

```python
fin_list = []
for i in range(len(list1)):
    if(list1[i]==1 and list2[i]==1):
        fin_list.append(1)
    else:
        fin_list.append(0)
pd.DataFrame(fin_list)[0].value_counts()
```

```python
fin_df = pd.concat([data2, pd.DataFrame(fin_list)], axis = 1)
```

```python
fin_df.head()
```

```python
fin_df.set_index("관광지명", inplace=True)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
```

```python
X = fin_df.iloc[:,:12]
target = fin_df.iloc[:,-1]
```

```python
X_train, X_test, target_train, target_test = train_test_split(X, target) 
print(X_train.shape)
print(X_test.shape)
print(target_train.shape)
print(target_test.shape)
```

```python
for k in range(1, 100): # k = 13
    knn = KNeighborsClassifier(n_neighbors=k)
    result = cross_val_score(knn, X_train, target_train, cv = 5)
    print(k, result.mean())
```

```python
for c in [0.0001, 0.001, 0.01,1, 0.1, 10, 100, 1000, 10000]:#c = 10
    lr = LogisticRegression(C=c)
    result = cross_val_score(lr, X_train, target_train, cv=5)
    print(c, result.mean())
```

```python
for d in range(1, 50): #d = 5
    model = DecisionTreeClassifier(max_depth = d)
    result = cross_val_score(model, X_train, target_train, cv = 5)
    print(d, result.mean())
```

```python
from sklearn.ensemble import VotingClassifier
final_lr = LogisticRegression(C=10)
final_tree = DecisionTreeClassifier(max_depth=7)
voting = VotingClassifier(estimators = [('lr', final_lr),('tree', final_tree)], voting='soft')
voting.fit(X_train, target_train)
print(voting.score(X_train,target_train))
print(voting.score(X_test, target_test))
```

```python
for n in range(1,50):
    forest = RandomForestClassifier(n_estimators=n)
    result = cross_val_score(forest, X_train, target_train, cv = 5)
    print(n, result.mean())
```
